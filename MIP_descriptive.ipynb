#MIP Analysis

This script analyses the data generated by the merger of the ownership data and the company panel information

```python
import pandas as pd
import numpy as np
```

Load the datasets

```python
df_merged_companies = pd.read_csv(r"C:\Users\lucas\OneDrive\BA\Data\outputs\merged_companies_reganalysis.csv")
df_merged_owners = pd.read_csv(r"C:\Users\lucas\OneDrive\BA\Data\outputs\merged_owners_reganalysis.csv")
```

```python
df_merged_companies["period"] = np.where(df_merged_companies["jahr"] <= 2008, "2006-2008", np.where((df_merged_companies["jahr"] >= 2009) & (df_merged_companies["jahr"] <= 2011), "2009-2011", np.where((df_merged_companies["jahr"] >= 2012) & (df_merged_companies["jahr"] <= 2014), "2012-2014", np.where((df_merged_companies["jahr"] >= 2015) & (df_merged_companies["jahr"] <= 2017), "2015-2017", np.where((df_merged_companies["jahr"] >= 2018) & (df_merged_companies["jahr"] <= 2020), "2018-2020", "NA")))))
df_merged_owners["period"] = np.where(df_merged_owners["jahr"] <= 2008, "2006-2008", np.where((df_merged_owners["jahr"] >= 2009) & (df_merged_owners["jahr"] <= 2011), "2009-2011", np.where((df_merged_owners["jahr"] >= 2012) & (df_merged_owners["jahr"] <= 2014), "2012-2014", np.where((df_merged_owners["jahr"] >= 2015) & (df_merged_owners["jahr"] <= 2017), "2015-2017", np.where((df_merged_owners["jahr"] >= 2018) & (df_merged_owners["jahr"] <= 2020), "2018-2020", "NA")))))
```

```python
print("Descriptive statistics for 'bges' variable in df_merged_companies for each 'period':")
display(df_merged_companies.groupby(['period', 'is_owner_change_2009_2017'])['bges'].describe())
print("Descriptive statistics for 'bges' variable in df_merged_owners for each 'period':")
display(df_merged_owners.groupby('period')['bges'].describe())
print("Descriptive statistics for 'emissions_CO2' variable in df_merged_companies for each 'period':")
display(df_merged_companies.groupby(['period', 'is_owner_change_2009_2017'])['emissions_CO2'].describe())
print("Descriptive statistics for 'emissions_CO2' variable in df_merged_owners for each 'period':")
display(df_merged_owners.groupby('period')['emissions_CO2'].describe())
```

Copy the environmental innovation data from 2008, 2014, and 2020 to the previous two years

```python
eco_innovations = ["oekpz1", "oekpz2", "oekpz3", "oekpz4", "oekpz5", "oekpz6", "oekpz7", "oekpz8", "oekpz9", "oekpz_avg", "oekpd1", "oekpd2", "oekpd3", "oekpd4", "oekpd_avg" , "emissions_CO2", "emissions_other" , "material_use", "recycling", "hazardous_material"]
for variable in eco_innovations:
    df_merged_companies[variable] = df_merged_companies.groupby(['companyid', 'period'])[variable].transform('max')
    df_merged_owners[variable] = df_merged_owners.groupby(['companyid', 'period'])[variable].transform('max')
```

```python
pd.crosstab(df_merged_companies['jahr'], df_merged_companies['emissions_CO2'], margins=True)
```
